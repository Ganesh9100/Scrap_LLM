## **simple crawling**

```python

import asyncio
import nest_asyncio
from crawl4ai import AsyncWebCrawler

nest_asyncio.apply()

async def main():
  async with AsyncWebCrawler() as Crawler:
    result = await Crawler.arun(url="https://www.straighttalk.com/support/terms-conditions")

    print(result)

if __name__ == '__main__':
  loop = asyncio.get_event_loop()
  loop.run_until_complete(main())
```

```python
import asyncio
import nest_asyncio
from crawl4ai import AsyncWebCrawler

nest_asyncio.apply()

async def main():
  async with AsyncWebCrawler(managed=True) as Crawler:
    result = await Crawler.arun(url="https://crawl4ai.com/mkdocs/")

    print(result.markdown)

if __name__ == '__main__':
  loop = asyncio.get_event_loop()
  loop.run_until_complete(main())
```

```python
import requests
from bs4 import BeautifulSoup
from jina import Flow, Document, DocumentArray
import os
os.environ["JINACHAT_API_KEY"] ="jina_f0c4924a6af348b9acadcdde14358dd41y87PYbPuv9hgY-VyJquRzppQ0Jx"
def scrape_websites(urls):
    docs = DocumentArray()
    for url in urls:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        text = soup.get_text()
        doc = Document(text=text, tags={'source': url})
        docs.append(doc)
    return docs
urls = [
    "https://www.straighttalk.com/support/terms-conditions"
    ]
documents = scrape_websites(urls)
flow = Flow().add(uses='jinahub://TextReader')
if __name__ == '__main__':
    with flow:
        flow.index(inputs=documents)
    def search_query(query):
        with flow:
            results = flow.search(inputs=Document(text=query))
            for doc in results:
                print(f"Source: {doc.tags['source']}")
                print(f"Text: {doc.text}")
    search_query("best and cheapest AI tools for software developers")
```
